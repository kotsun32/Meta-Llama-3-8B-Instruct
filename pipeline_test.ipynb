```python
import torch
import transformers

class ChatBot:
    def __init__(self, model_id, device="cuda", dtype=torch.bfloat16, system_message="You are a helpful assistant"):
        self.model_id = model_id
        self.device = device
        self.dtype = dtype
        self.pipeline = transformers.pipeline(
            "text-generation",
            model=self.model_id,
            model_kwargs={"torch_dtype": self.dtype},
            device=self.device,
        )
        self.history = []
        self.system_message = system_message
        self.history.append({"role": "system", "content": system_message})

    def add_to_history(self, role, content):
        message = {
            "role": role,
            "content": content,
        }
        self.history.append(message)

    def generate_response(self, prompt, max_tokens=500):
        # Add user prompt to history
        self.add_to_history(role="user", content=prompt)
        
        # Generate the conversation history string without repeating the system message
        conversation_history = f"system: {self.system_message}\n" + "\n".join([f"{msg['role']}: {msg['content']}" for msg in self.history if msg['role'] != 'system']) + "\nassistant:"

        # Generate response within token limit
        response = self.pipeline(conversation_history, max_length=max_tokens, num_return_sequences=1, pad_token_id=50256)[0]["generated_text"]
        
        # Extract the assistant's response from the generated text
        assistant_response_start = response.find('assistant:') + len('assistant:')
        assistant_response_end = response.find('user:', assistant_response_start)
        
        if assistant_response_end == -1:
            assistant_response = response[assistant_response_start:].strip()
        else:
            assistant_response = response[assistant_response_start:assistant_response_end].strip()
        
        # Add assistant's response to history
        self.add_to_history(role="assistant", content=assistant_response)
        
        return assistant_response
    
    def prompt_user(self, prompt):
        response = self.generate_response(prompt)
        print("Assistant:", response)


```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

class ChatBot:
    def __init__(self, model_id, device="cuda", dtype=torch.bfloat16, system_message="You are a helpful assistant"):
        self.model_id = model_id
        self.device = torch.device(device if torch.cuda.is_available() else "cpu")
        self.dtype = dtype
        self.tokenizer = AutoTokenizer.from_pretrained(self.model_id)
        self.model = AutoModelForCausalLM.from_pretrained(self.model_id, torch_dtype=self.dtype).to(self.device)
        self.pipeline = pipeline(
            "text-generation",
            model=self.model,
            tokenizer=self.tokenizer,
            device=0 if self.device.type == "cuda" else -1,
        )
        self.history = [{"role": "system", "content": system_message}]

    def add_to_history(self, role, content):
        message = {
            "role": role,
            "content": content,
        }
        self.history.append(message)

    def generate_response(self, prompt, max_tokens=500):
        # Add user prompt to history
        self.add_to_history(role="user", content=prompt)
        
        # Generate the conversation history string
        conversation_history = "\n".join([f"{msg['role']}: {msg['content']}" for msg in self.history]) + "\nassistant:"

        # Encode the conversation history
        input_ids = self.tokenizer.encode(conversation_history, return_tensors="pt").to(self.device)

        # Generate response within token limit
        response_ids = self.model.generate(
            input_ids, 
            max_length=input_ids.shape[1] + max_tokens, 
            pad_token_id=self.tokenizer.eos_token_id,
            no_repeat_ngram_size=3,  # to reduce repetitive responses
            num_beams=5,  # to consider multiple beams and select the best
        )
        
        # Decode the generated response
        response = self.tokenizer.decode(response_ids[0], skip_special_tokens=True)
        
        # Extract the assistant's response from the generated text
        assistant_response_start = response.find('assistant:') + len('assistant:')
        assistant_response_end = response.find('user:', assistant_response_start)
        if assistant_response_end == -1:
            assistant_response = response[assistant_response_start:].strip()
        else:
            assistant_response = response[assistant_response_start:assistant_response_end].strip()
        
        # Add assistant's response to history
        self.add_to_history(role="assistant", content=assistant_response)
        
        return assistant_response
    
    def prompt_user(self, prompt):
        response = self.generate_response(prompt)
        print("Assistant:", response)


```
